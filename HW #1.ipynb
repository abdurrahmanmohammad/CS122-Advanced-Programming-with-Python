A N-layer tree  had layers labeled from 0,1,2, up n.
* Each layer i is a trapezoid with the 
    * bottom width = 2*hw+1,   where hw = i*widen_scale
    * height       = (hw+1)//2 + 1
    * top width    = (bottom width) - 2*(height) + 2
* The tree trunk is a rectangle of letters "T" with 
    * width        = (n//3)*2+1
    * height       = 2*width
* Each layer i is labeled with  digit i.
