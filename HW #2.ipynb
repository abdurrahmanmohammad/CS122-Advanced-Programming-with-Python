```python
class AsciiObj:
    SPACE = " "          # the default space character
        
    @staticmethod        
    def addBorder(lines, bg=None, k=1):
        """The is a static method 
            * The input argument lines represent objects with background symbol bg
            * If k > 0, then it replace k-layer border with symbol "@" around the object represented in lines
            * This is a helper function for drawLines
        """
        if bg == None or k <=0:
            return 
        m = len ( lines)
        n = len (lines[0])
        mark = "@"
        for i,s in enumerate (lines): # for row s=lines[i]
            s2 = ""
            for j, x in enumerate (s):  # for character x=s[j]
                if x != bg:             # if the current character x is not background symbol
                    if  (i<k or i>=m-k or j<k or j>=n-k):   # xxx corrections
                        # check if x is within the k layers of the screen borders  (left,right, top, bottom)
                        x = mark         # replace x with special symbol if yes                     
                    else: # xxx corrections here
                        for a in range (1,k+1):
                            found = lines[i-a][j]==bg or lines[i+a][j]==bg or s[j-a] == bg or s[j+a] ==bg
                            if found:
                                x = mark
                                break;
                s2 += x
            lines[i] = s2
    @staticmethod        
    def  rotate90 (lines):
        """Rotate the object represented by "lines" 90 degree clockwise
        * This is a static method to be shared with class Tree, Forest

        """
        m = len (lines)
        n = len (lines[0])
        ans = []
        for i in range (n):
            x=""
            for s in lines[::-1]:
                x += s[i]
            ans.append (x)
        return ans

    @staticmethod        
    def drawLines (lines, bg=None, reverse=False, margin=0, border=0): 
        """* Print out lines with 
            * the number of spaces to the left screen border = margin (default 0)
            * If border=n > 0, then replace the n layers of the object borders with special symbol @.
            * If reverse=True, the objects will be printed out upside down
            * This is a static method to be used by both class Tree and class Forest
        """           
        if reverse:
            lines.reverse() 

        if border > 0 :
            AsciiObj.addBorder (lines, bg, border)

        offset = Tree.SPACE * margin
        for s in lines:
            print (offset+s)
```

```python
# Code copied from HW#1 solution
class Tree:
    SPACE = " "
    FG = "*"
    NEWLINE = "\n"
    """
    """
    def __init__(self, layers=7, widen = 1, margin=0, bg= SPACE):
        self.lines = []
        self.bg = bg
        self.layers = layers
        self.widen = 0
        self.margin = margin
        self.build (widen, bg)
        self.rotation = 0
    
    def rotate (self):
        """Roate the tree 90 degree clockwise
        """
        self.rotation = (self.rotation + 1) % 4
        self.lines = AsciiObj.rotate90(self.lines)

    # Revise the following function as follows:
    # 1. add new parameter border with default 0
    # 2. change the default of bg to None
    def draw (self, reverse=False, bg=None, margin=0, border=0):  
        """Print out the list of strings that represents the tree with
            * background symbol=bg (default None )
            * the number of spaces to the left screen border = margin (default 0)
            * If border=n > 0, then replace the n layers of the tree border with special symbols.
            * If reverse=True, the tree will be printed out upside down """
        if bg:
            for i, s in enumerate(self.lines):
                self.lines[i] = s.replace(self.bg, bg)
            self.bg = bg
        AsciiObj.drawLines (self.lines[:], bg, reverse, margin, border)

    
        
    def line(self, x, n, fg=FG): # make line of n chararcters ch center at position x
        """return a string of n chararcters fg center at position x
        """
        bg = self.bg
        spaces = self.margin*bg
        side = bg*(x-n//2)
        #x = fg*(n//2) + "0"+ fg*(n//2) 
        x = fg*n
        return side+ x + side +spaces
    
    def triangle (self, x, hw, fg=FG):
        """Return a list of strings that represent a trapezoid with 
            * center x-coordinate = x
            * bottom width = 2*hw + 1
            * height = (hw+1)//2 + 1
            * top width    = (bottom width) - 2*(height) + 2""" 
        s = []
        for i in range (hw//2, hw+1):
            length = 2*i+1
            tmp = self.line(x,length, fg) 
            s.append (tmp)
        return s
    def treeTop (self, x):
        """Return a list of strings that represent n-layered trapzoid of the tree where
                * n = self.
        """
        s = []
        for i in range (self.layers+1):
            j = i%10
            hw = i*self.widen 
            s += self.triangle(x,hw, str(j) )
        return s
    def treeTrunk (self, x,hw):
        """Return a list of strings that represent the tree trunk with
            * width = 2*hw+1
            * height = 2*width
        """
        s = [] 
        w = 2*hw + 1
        h = 2*w
        for i in range(h):
            s.append ( self.line (x, w, "T") )
        return s

    def build (self, widen=1, bg=SPACE):
        """Compose a list of lines that represent a n-layer tree with
            * the background symbol = bg  (default is the space character)
            * the tree width = 2*n*widen+1 (default 1)"""
        if widen <= 0 or widen==self.widen:
            return
        self.widen = widen
        hw = (self.layers)*self.widen
        x  = hw  
        self.lines =  self.treeTop (x)
        self.lines += self.treeTrunk (x,hw//3)
    
    def __len__(self):
        """Return the height of the tree.
        """
        return len (self.lines)
    
    def getLines (self):
        """Return a list of strings that represents the tree
        """
        return self.lines
```

```python
class Forest():

    def __init__ (self, ntree = [3,7,1,9], widen=1, margin=0, bg=Tree.SPACE):
        """Constructor: build trees of layers specified in the list ntree. (The default is 4 trees of 3-, 7-, 1-, 9- layers.)
            * self.lines is a list of strings use to draw a forest in ASCII
        """
        # Initialize class variables
        
        # Variables also in Tree 
        self.lines = [] # Stores the lines to print the forest
        self.bg = bg
        self.widen = 0
        self.margin = margin
        
        # Forest variables
        self.ntree = ntree[:] # Number of trees in the forest
        self.trees = [] # Stores the trees in the forest
        self.tallest = 0 # Keeps track of the tallest tree in the forest
        self.rotation = 0 # Keeps track of forest rotation (1 to 4)
        
        self.build(widen, bg) # Build forest
        

    def setLines(self): # you may or may not implement this funcition
        """Compose a list of strings that represents the forest by combining the strings that 
            represent each individual tree.
           This is a helper function for build and rotate """
        self.lines = [] # Clear existing lines
        self.tallest = 0 # Clear existing tallest (tree)
        # Determine tallest tree
        for t in self.trees: # Iterate over all trees
            self.tallest = max(self.tallest, len(t)) # Compare tree length with current max until get tallest tree
        # Add upper padding in forest
        tallest = self.tallest
        for i in range(tallest):
            s = "" # Stores line
            for t in self.trees: # Iterate over trees in forest
                tree = t.getLines() # Get lines of each tree
                n = len(tree) # Get length of list (list containing the lines of the tree)
                m = len(tree[0]) # Get length of each line stored in list
                j = i - (tallest - n)
                if 0 <= j:
                    s += tree[j]
                else:
                    s += self.bg * m # Upper pad the string
            self.lines.append(s) # Add line to forest
    
    def build(self, widen, bg=Tree.SPACE):
        # Set instance variables
        self.widen = widen
        self.bg = bg
        self.trees = [] # Reset trees everytime we build and rebuild forest
        # Make the trees
        for i in self.ntree: # Iterate over the number of trees
            tree = Tree(i, self.widen, self.margin, self.bg) # Create a tree
            self.trees.append(tree) # Add the tree to the forest
        self.setLines() # Set lines
            
    # Note: Change the default of bg to None   
    def draw(self, reverse=False, bg=None, margin = 0, border=0 ):
        """Print out the list of strings that represents the forest with
            * background symbol=bg --> default was changed to None
            * the number of spaces to the left boarder of the window = margin (default 0)
           If reverse=True, the forest will be printed out upside down 
           If border>0, then the border of the objects will be replaced with the special symbols. 
        """
        if bg:
            for i, s in enumerate(self.lines): # line = self.lines[i]
                self.lines[i] = s.replace(self.bg, bg) # Replace old bg char with new bg char
            self.bg = bg # Update background char (after using old and new bg char in loop)
        AsciiObj.drawLines(self.lines[:], bg, reverse, margin, border) # Get the new forest lines
           
    def  rotate (self, individual_rotation=None):
        """Revise the list of strings self.lines that represents the rotated trees.
            if individual_rotation=k is specified, then tree k is rotated 90 degree clockwise individually.
            If individual_rotation=None, then the entire forest will be rotated 90 degree clockwise.
        """
        # Check if no param first (type None) or error in 2nd condition
        if not individual_rotation: # If no param, change forest rotation variable
            self.rotation = (self.rotation + 1) % 4 # Update rotation variable
        elif 0 <= individual_rotation and individual_rotation < len(self.trees): # Rotate a certain tree
            self.trees[individual_rotation].rotate() # Rotate the tree indicated by individual_rotation
        self.setLines() # Set the lines
        for i in range(self.rotation): # Using forest rotation variable, rotate forest
            self.lines = AsciiObj.rotate90(self.lines) # Rotate forest
```
