```python
import requests # Same as week 12
import pandas as pd
web = "http://www.metoffice.gov.uk/pub/data/weather/uk/climate/stationdata/heathrowdata.txt" # Same as week 12
response = requests.get(web) # Same as week 12
data = response.text # Same as week 12
data_rows = [] # Same as week 12
#rainfall = [] # Same as week 12
#maximumTemperature = [] # for maximum temperatures
#minimumTemperature = [] # for minimum temperatures

years = set() # Set {year0, year1, year2, ...}: Stores the years of each row
tmax = {} # Dictionary tmax[year] = val: Stores the tmax of each row
tmin = {} # Dictionary tmin[year] = val: Stores the tmin of each row
rain = {} # Dictionary rain[year] = val: Stores the rain of each row

for row in data.split("\r\n")[7:]: # Same as week 12
    fields = [x for x in row.split(" ") if x] # Same as week 12
    data_rows.append(fields) # Same as week 12
    
    # Values for the dataset as a whole: Avg rainfall, highest temparature, lowest temparature
    #rainfall.append(float(fields[5])) # Same as week 12
    #maximumTemperature.append(float(fields[2])) # Store maximum temperature for the entry
    #minimumTemperature.append(float(fields[3])) # Store minimum temperature for the entry
    
    # Extract and store year
    year = int(fields[0]) # Get the year (at index 0)
    years.add(year) # Store the year
    
    # Create a dictionary entry for each year (if not already created)
    tmax.setdefault(year,[]) 
    tmin.setdefault(year,[])
    rain.setdefault(year,[])
    
    # Store data in dictionaries for each row
    tmax[year].append(float(fields[2])) # Store tmax of all rows for a particular year
    tmin[year].append(float(fields[3])) # Store tmin of all rows for a particular year
    rain[year].append(float(fields[5])) # Store rain of all rows for a particular year

#print("Average rainfall = {:.1f} mm".format(sum(rainfall) / len(rainfall))) # Print average rainfall
#print("The highest temparature in degree = {}".format(max(maximumTemperature))) # Print maximum temperature
#print("The lowest temparature in degree = {}".format(min(minimumTemperature))) # Print minimum temperature

# Calculations
years = list(years) # Convert set to list
highest_degree = {} # Stores the maximum temperature for the year (maximum of tmax entries for each year)
lowest_degree = {} # Stores the minimum temperature for the year (minimum of tmin entries for each year)
rain_fall = {} # Stores the avg rainfall for each year (sum rainfall of each entry and divide by length)

for year in years: # Iterate over each unique year
    highest_degree[year] = max(tmax[year]) # Find the maximum tmax entry per year
    lowest_degree[year] = min(tmin[year])  # Find the minimum tmin entry per year
    # Calculate avg rainfall per year. Round to 1 decimal place.
    rain_fall[year] = round(sum(rain[year]) / len(rain[year]), 1)

# Convert dictionaries to list
rain_fall = list(rain_fall.values()) 
highest_degree = list(highest_degree.values())
lowest_degree = list(lowest_degree.values())

print("Average rainfall = {:.1f} mm".format(sum(rain_fall) / len(rain_fall))) # Total avg rainfall for the dataset
print("The highest temparature in degree = {}".format(max(highest_degree))) # Maximum tmax of dataset
print("The lowest temparature in degree = {}".format(min(lowest_degree))) # Minimum tmin of dataset

frame = {'Year':years,'Highest_Degree':highest_degree,'Lowest_Degree':lowest_degree,'Rainfall':rain_fall}
table = pd.DataFrame(frame)
print(table.to_string(index = False,justify='center')) # Print the table
```

# Q2: (15pt) Web Scraping with Python Using Beautiful Soup

```python
import requests
from bs4 import BeautifulSoup
web = 'https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168'
response = requests.get(web)
soup = BeautifulSoup(response.text,'html.parser')
seven_day = soup.find(id = "seven-day-forecast")
forecast_items = seven_day.find_all(class_ = "tombstone-container")
weathers = seven_day.find_all(class_ = 'short-desc')

weather = [x.text for x in weathers]
conditions = []
periods = []

for i in range(len(weather)):
    if weather[i] not in conditions:
        conditions.append(weather[i])
conditions.sort()
conditions.insert(0, '_'*13)

for i in range(1, len(conditions)):
    periods.append(weather.count(conditions[i]))
periods.insert(0,'_'*13)

import pandas as pd
frame = {'Weather':conditions,'# Periods':periods}
table = pd.DataFrame(frame)
print(table.to_string(index = False))
# I could have iterated the output and left-justified, but using the to_string was more convenient
```
